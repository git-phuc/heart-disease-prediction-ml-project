# 7 — Demo inference (load .joblib và dự đoán)

Notebook này minh hoạ cách load model đã lưu trong `saved_models/` và chạy dự đoán.

**Phụ thuộc:** đã chạy `app/6_test_and_save.ipynb` để có file `.joblib`.


```python
from pathlib import Path
import joblib
import pandas as pd

PROJECT_ROOT = Path.cwd()
MODEL_PATH = PROJECT_ROOT / "saved_models" / "CHD_RF_tuned.joblib"  # hoặc CHD_ET_tuned.joblib
DATA_PATH = PROJECT_ROOT / "data" / "D11KS.csv"

bundle = joblib.load(MODEL_PATH)
pipe = bundle["pipeline"]
thr  = bundle.get("threshold", 0.5)

df = pd.read_csv(DATA_PATH)
TARGET = "TenYearCHD"
X = df.drop(columns=[TARGET], errors="ignore")

proba = pipe.predict_proba(X)[:, 1]
pred = (proba >= thr).astype(int)

print("Model:", bundle.get("name","?"), "| thr:", thr)
print("Pred sample:", pred[:10].tolist())
print("Proba sample:", [float(x) for x in proba[:10]])
```
